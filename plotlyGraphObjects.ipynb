```python
anames=a90.index.to_list() 
ahamcha=ipl[ipl['batter'].isin(anames)] 
ahamcha 
#ahamcha is used for calulating runs
```

```python
ahamchas_rate=ahamcha[~(ahamcha['extras_type'].isin(['wides']))] 
ahamchas_rate
#ahamchas_rate for calculating the number of balls
```

```python
#ahamcha and ahamchas_rate 
runs=ahamcha.groupby('batter')['batsman_runs'].sum().sort_values(ascending=False)
runs 
balls=ahamchas_rate.groupby('batter')['batsman_runs'].count() 
balls 
strike_rate_per_batsmen=((runs/balls)*100).sort_values(ascending=False) 
strike_rate_per_batsmen=strike_rate_per_batsmen.reset_index() 
strike_rate_per_batsmen.replace({'batsman_runs':'batsman_strikerate'},inplace=True) 
strike_rate_per_batsmen
```

```python
#I want to plot a graph that would take any number of arguments and build a graph on it. The arguments would contain names of 
#batters. 
import plotly.offline as pyo 
import plotly.graph_objs as go 

def build_line_graphs(*names):
    data=[] 
    for i in names:
        uq=ipl[ipl['batter']==i] 
        a1=uq.groupby('season')['batsman_runs'].sum().reset_index() 
        trace=go.Scatter(x=a1['season'],y=a1['batsman_runs'],mode='lines+markers',name=i)
        data.append(trace) 
    layout=go.Layout(title='Season Vs Runs',
                    xaxis={'title':'Seasons'},
                    yaxis={'title':'Runs'}) 
    fig=go.Figure(data,layout) 
    pyo.plot(fig,filename='mujydre.html')
```

```python
import plotly.offline as pyo
import plotly.graph_objs as go
import pandas as pd

def build_line_graphs(*names):
    data = []
    seasons = sorted(ipl['season'].unique())  # Ensure all seasons are covered
    
    for batter_name in names:
        # Filter the data for the specific batter
        batter_data = ipl[ipl['batter'] == batter_name]
        
        # Group by season and sum the runs
        season_runs = batter_data.groupby('season')['batsman_runs'].sum().reset_index()
        
        # Reindex to ensure all seasons are included, even if a batter has no runs in some seasons
        season_runs = season_runs.set_index('season').reindex(seasons, fill_value=0).reset_index()
        
        # Create the line trace with distinct markers and line styles
        trace = go.Scatter(
            x=season_runs['season'], 
            y=season_runs['batsman_runs'], 
            mode='lines+markers', 
            name=batter_name   # Customize line width
        )
        data.append(trace)
    
    # Define the layout with titles and axis labels
    layout = go.Layout(
        title='Season Vs Runs of Batters',
        xaxis={'title': 'Seasons'},
        yaxis={'title': 'Runs'},
        hovermode='closest'  # Ensure hover shows relevant points clearly
    )
    
    # Create the figure with data and layout
    fig = go.Figure(data=data, layout=layout)
    
    # Plot the graph
    pyo.plot(fig, filename='batters_season_vs_runs.html')
```
